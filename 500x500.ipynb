```
from itertools import chain
from collections import defaultdict
import random
import string
import pandas as pd
import csv
import codecs




csv_colums = ['회원번호','회원이름','나이','성별','관심분야']



# 가장 많은 성 10개 중 한 개를 선별 이름엔 들어갈 땐 랜덤으로 설정됨.
first_name = ['김','이','박','최','정','강','조','윤','장','임']
name= string.ascii_uppercase

# csv 파일 오픈, 덮어쓰기를 위해 'w'를 사용
# newline은 줄 바꿈을 위해 사용
# 한글 깨짐 방지를 위해 encoding euc-kr 사용
with open('tuser1.csv','w',newline='',encoding='euc-kr') as f:
  writer = csv.DictWriter(f, fieldnames=csv_colums)
  writer.writeheader()
# 회원 테이블을 만들기 위해 Dictionary 형태로 만들어줌(test : 100명)
  for i in range(1,3001) :  
      result = "" # 결과 값
      for a in range(1,4) :
        if(a == 1) :
             result += random.choice(first_name) # 랜덤한 문자열 하나 선택
        else :
             result += random.choice(string.ascii_lowercase)
            
    # ID,이름,나이,성별,관심분야에 랜덤 데이터 값 저장    
      dict1 = {'회원번호': i,'회원이름': result, '나이': random.randrange(1,70), '성별': random.randrange(0,2), '관심분야':random.randrange(0,9) }
    # 한개씩 밑으로 저장
      writer.writerow(dict1)
              

tuser = pd.read_csv('tuser1.csv',encoding = 'euc-kr')
tuser1 = tuser[['회원번호','회원이름','나이','성별','관심분야']]
tuser1['대출건수'] = 0
tuser1
```

```
import pandas as pd
# 도서 대출건수 상위 30위 데이터 저장
tbook = pd.read_csv('libtest(500).csv',encoding = 'cp949')
tbook1 = tbook[['ISBN','도서명','저자','주제분류번호','대출건수']]
tbook1
```

```
from pandas import Series,DataFrame
import random
#회원과 도서테이블의 대출 건수와 매칭시키는 반복문 ex) 회원번호 1 + 도서

# ※ DataFrame에서 한 행만 추출할 경우 Series의 형태로 추출됨. 즉 DataFrame (회원,도서) -> Series -> DataFrame(대여목록) 으로 진행함.
# 각 DataFrame에서 추출해 저장할 Series의 틀과, 결과값으로 저장될 DataFrame의 틀을 만듦
tseries = ['회원번호','회원이름','나이','성별','관심분야','ISBN','도서명','저자','주제분류번호']
result1 = pd.DataFrame(columns = ['회원번호','회원이름','나이','성별','관심분야','ISBN','도서명','저자','주제분류번호'])

# 도서 매칭 반복문 
# a는 인덱스 번호선언
a = 0
# 30권책에서 한 권을 추출
for i in range(0,500) :
  # 한 권의 데이터를 Series형태 저장
  ltest1 = Series(tbook1.iloc[i],index=tseries)
  #추출한 책의 대출 건수만큼 반복문을 돌림
  for j in range(0,int(tbook1.loc[i,['대출건수']])) :
    unum =int(random.randrange(0,3000))
    # 200명의 회원중 한명을 rtest1(Series)에 저장
    rtest1 = Series( tuser1.iloc[unum],index=tseries)
    # 뽑힌 회원의 대출건수 더하기
    tuser1.iloc[unum,5] += 1
    # rtest1를 DataFrame 형태로 변환
    rtest2 = rtest1.to_frame().T
    # 최종 결과물인 result1(Dataframe)에 rtest2를 추가 
    result1 = result1.append(rtest2,ignore_index=True)
    # 이게뭐지..?
    result1=  result1.reset_index(drop=True)
    # Itest 책의 정보를 result1에 속성에 맞게 저장
    result1.set_value(a,'ISBN',ltest1['ISBN'])
    result1.set_value(a,'도서명',ltest1['도서명'])
    result1.set_value(a,'저자',ltest1['저자'])
    result1.set_value(a,'주제분류번호',ltest1['주제분류번호'])
    # set_value에 사용할 a(인덱스) 증가
    a = a+1
result1
```
