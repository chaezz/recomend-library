```python
# 유저테이블을 대출건수 기준으로 내림차순으로 정렬(Top 10,50,100 읽을 수 있음)
tuser3 = tuser2.sort_values(["대출건수"], ascending=[False])
# 50명의 대출기록을 빼는게 1950명의 대출기록을 빼는방법밖에 못찾음
tuser4 = tuser3.index[500:]

tuser4 = tuser4 + 1

tuser4



```

```python
# user 500위 이상 기록제거
result1 = result
for liuser in tqdm(range(0,len(tuser4)), desc = "진행상황"):  
        result1 = result1[result1.회원번호 != tuser4[liuser]]       
result1

```

```python
result1.to_csv("Rentlist(user500).csv", mode = 'w', encoding = 'cp949')
matrix = pd.pivot_table(result1,index='회원번호', columns='ISBN', values='value',fill_value = 0)
matrix.shape
```

```python
matrix = pd.pivot_table(result1,index='회원번호', columns='ISBN', values='value',fill_value = 0)
matrix.shape
```

```python
itemsum = matrix.sum(axis = 0, skipna = True) 
```

```python
itemsort = itemsum.sort_values(ascending=False)
```

```python
itemdf = itemsort.to_frame()
```

```python
outitem = itemdf.index[500:]
```

```python
# item 대출건수기준 500위 이상 기록제거
result2 = result1
for liitem in tqdm(range(0,len(outitem)), desc = "진행상황"):  
        result2 = result2[result2.ISBN != outitem[liitem]]       
result2

```

```python
matrix500 = pd.pivot_table(result2,index='회원번호', columns='ISBN', values='value',fill_value = 0)
```

```python
matrix500.to_csv('finalmatrix(500).csv',encoding = 'cp949')
```
